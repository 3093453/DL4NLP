```python
from deep_translator import GoogleTranslator

source = 'de'

langs = [
    #'en',
    #'da',
    'cs',
    #'pl'
]

for lang in langs:
    print(lang)
    file = f"unclean-{source}-{lang}.csv"
    path = "/".join( (cwd, folder, file) )
    
    df = pd.read_csv(path, index_col=0)

    translator = GoogleTranslator(source=lang, target=source)
    #translator = DeeplTranslator(api_key, source=source, target=lang)

    translations = translator.translate_batch(df.Textstelle.to_list())
    
    data = df.copy()
    data["Textstelle"] = translations
    
    file = f"unclean-{lang}-{source}.csv"
    path = "/".join( (cwd, folder, file) )

    data.to_csv(path)

# en, da, cs, pl
```
